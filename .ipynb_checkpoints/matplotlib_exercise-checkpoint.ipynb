## Matplotlib Assignment

```python
# import pandas
import pandas as pd

# import matplotlib
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:\\Users\\bevli\\Downloads\\company_sales_data.csv")
```

```python
df.head()
```

### Task 1: Read the `total_profit` of all months and display it using a line plot.

```python
profit = df['total_profit']
months = df['month_number']
plt.plot(months,profit,label= 'Monthwise plot of profits')
plt.xlabel('Month number of year')
plt.ylabel('Profit in $')
plt.show()
```

```python
#Stateless Approach
profit = df['total_profit']
months = df['month_number']
fig, ax = plt.subplots(figsize = (10,6))
ax.plot(months, profit, label = "Monthwise plot of profits")
ax.set_xlabel("Month")
ax.set_ylabel("Profit")
ax.set_title("Monthly profits")
```

### Task 2: Read all the different product sales data and display it using a multiline plot.

```python
facecream = df['facecream']
facewash = df['facewash']
toothpaste = df['toothpaste']
bathingsoap = df['bathingsoap']
shampoo = df['shampoo']
moisturizer = df['moisturizer']
months = df['month_number']

plt.plot(months, facecream,   label = 'Face cream Sales Data')
plt.plot(months, facewash,   label = 'Facewash Sales Data')
plt.plot(months, toothpaste,   label = 'Toothpaste Sales Data')
plt.plot(months, bathingsoap,   label = 'Bathing Soap Sales Data')
plt.plot(months, shampoo,   label = 'Shampoo Sales Data')
plt.plot(months, moisturizer,   label = 'Moisturizer Sales Data')
plt.legend(loc='upper right')
plt.xlabel('Month number of year')
plt.ylabel('Product sales')
plt.title('Product Sales')
plt.show()
```

```python
#stateless approach
facecream = df['facecream']
facewash = df['facewash']
toothpaste = df['toothpaste']
bathingsoap = df['bathingsoap']
shampoo = df['shampoo']
moisturizer = df['moisturizer']
months = df['month_number']

fig, ax = plt.subplots(figsize = (10,6))

ax.plot(months, facecream,   label = 'Face cream Sales Data')
ax.plot(months, facewash,   label = 'Facewash Sales Data')
ax.plot(months, toothpaste,   label = 'Toothpaste Sales Data')
ax.plot(months, bathingsoap,   label = 'Bathing Soap Sales Data')
ax.plot(months, shampoo,   label = 'Shampoo Sales Data')
ax.plot(months, moisturizer,   label = 'Moisturizer Sales Data')
ax.legend(loc='upper right')
ax.set_xlabel('Month number of year')
ax.set_ylabel('Product sales')
ax.set_title('Product Sales')
plt.show()
```

### Task 3: Read the `total_profit` of each month and display it using a histogram to find out which profit ranges are the most common.

```python
profit = df['total_profit']
months = df['month_number']
n,bins,patches = plt.hist( profit, label = "Profits Monthly", bins=10)
plt.xticks(bins)
plt.xlabel('Profits')
plt.ylabel('Counts')
plt.title('Profit Histogram')
plt.show()
```
